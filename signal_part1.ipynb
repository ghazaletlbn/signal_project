```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import irfft
from scipy import signal
from scipy.fft import fft, fftfreq
from scipy.fft import rfft, rfftfreq
from scipy import signal
from scipy.io import wavfile
```

```python
#bakhsh1
samplingFrequency, signalData= wavfile.read('/content/gdrive/My Drive/signal/CA3/noisy_song.wav')
plt.plot(signalData[0:1000])
plt.ylabel("signal")
plt.xlabel("Time")
plt.show()
```

```python
#bakhsh2
plt.specgram(signalData,Fs=samplingFrequency)
plt.ylabel("frequency")
plt.xlabel("Time")
plt.show()
```

```python
#bakhsh3
n = len(signalData)
yf = rfft(signalData)
xf = rfftfreq(n, 1 / samplingFrequency)

plt.plot(xf, np.abs(yf))
plt.show()
```

```python
#bakhsh3
points_per_freq = len(xf)/(samplingFrequency/2)
target_idx = int(points_per_freq*5000)
yf2 = yf
yf[target_idx-1:target_idx+2]=0
plt.plot(xf, np.abs(yf))
plt.show()
```

```python
#bakhsh4
new_sig = irfft(yf)
plt.subplot(1,2,1)
plt.plot(new_sig[:1000])
plt.show()
plt.subplot(1,2,2)
plt.specgram(new_sig,Fs=samplingFrequency)
plt.show()
wavfile.write("lowpass.wav",samplingFrequency,new_sig  )
```

```python
yf3 =yf2 -  yf
new_sig2 = irfft(yf3)
plt.subplot(1,2,1)
plt.plot(new_sig2[:1000])
plt.show()
plt.subplot(1,2,2)
plt.specgram(new_sig2,Fs=samplingFrequency)
plt.show()
wavfile.write("highpass.wav",samplingFrequency,new_sig2  )
```
