```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def conv2D(image, filters, stride, padding):
  f,f2,number_of_filters = filters.shape
  m1,n1,c1 = image.shape
  if(padding =='same'):
      new_image =np.zeros((m1+f-1, n1+f-1 ,c1))
      for c in range(0,c1):
          new_image[:,:,c] =np.pad(image[:,:,c],(f-1)//2,mode='constant')
      image = new_image
  m,n,c = image.shape
  f1 = (m-f)//stride
  f2 = (n-f)//stride
  output = np.zeros((f1+1, f2+1 ,number_of_filters))
  for final_c in range(0,number_of_filters):
      for row in range(0, f1+1):
          for col in range(0, f2+1):
              output[row, col,final_c] =0
              for c2 in range(0,c):
                  output[row, col,final_c] += np.sum(image[row:row+f, col:col+f,c2] * filters[:,:,final_c])

  return output
```

```python
horizontal_edge = np.array([[-1,-1,-1],
                   [0,0,0],
                   [1,1,1]])

vertical_edge = np.array([[-1,0,1],
                 [-1,0,1],
                 [-1,0,1]])

sharpening = np.array([[0,-1,0],
              [-1,5,-1],
              [0,-1,0]])

weighted_averaging_3x3 = (1/16)*np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]])
l1_filters = np.zeros((3,3,4))
l1_filters[:,:,0] = vertical_edge
l1_filters[:,:,1] = horizontal_edge
l1_filters[:,:,2] = sharpening
l1_filters[:,:,3] = weighted_averaging_3x3
```

```python
robert_x =np.array([[1, 0],
                    [0, -1]] )

robert_y =np.array([[0, +1],
                    [-1, 0]] )

averaging_2x2 = (1/4)*np.array([[1, 1],
                                [1, 1]] )
l2_filters = np.zeros((2,2,3))
l2_filters[:,:,0] = robert_x
l2_filters[:,:,1] = robert_y
l2_filters[:,:,2] = averaging_2x2
```

```python
#image 1 => layer1
im = cv2.imread('/content/gdrive/My Drive/signal/CA3/images/bengio.jpeg')
layer_one_output = conv2D(im, l1_filters,1,'same')
plt.subplot(3,2,1)
plt.imshow(im)
plt.subplot(3,2,2)
plt.imshow(layer_one_output[:,:,0])
plt.subplot(3,2,3)
plt.imshow(layer_one_output[:,:,1])
plt.subplot(3,2,4)
plt.imshow(layer_one_output[:,:,2])
plt.subplot(3,2,5)
plt.imshow(layer_one_output[:,:,3])
```

```python
#image 1 => layer2
layer_two_output = conv2D(layer_one_output, l2_filters,2,'valid')
plt.subplot(2,2,1)
plt.imshow(layer_one_output)
plt.subplot(2,2,2)
plt.imshow(layer_two_output[:,:,0])
plt.subplot(2,2,3)
plt.imshow(layer_two_output[:,:,1])
plt.subplot(2,2,4)
plt.imshow(layer_two_output[:,:,2])
```
